```python
import os
import glob
import yaml
import pandas as pd

# Folder where all YAML files are stored
folder = r"D:\stock analysis project final\venv\Scripts\raw"

# Pattern to match all YAML files in that folder (and subfolders if needed)
pattern = os.path.join(folder, "**", "*.yaml")

# List all YAML file paths
yaml_files = glob.glob(pattern, recursive=True)

all_records = []

for path in yaml_files:
    with open(path, "r", encoding="utf-8") as f:
        content = yaml.safe_load(f)

    # content can be:
    # - a list of dicts
    # - a single dict
    # Adjust based on your file structure
    if isinstance(content, list):
        all_records.extend(content)
    elif isinstance(content, dict):
        all_records.append(content)

# Convert to DataFrame
df = pd.DataFrame(all_records)

print(df.head())
print("Total rows:", len(df))
```

```python
df.to_csv('allstocks.csv', index=False)
```

```python
import pandas as pd
import os

# Read the CSV file
df = pd.read_csv('allstocks.csv')

# Group by 'symbol' and save each ticker to a separate CSV
output_dir = 'ticker_files'
os.makedirs(output_dir, exist_ok=True)

for symbol in df['symbol'].unique():
    ticker_df = df[df['symbol'] == symbol].copy()
    ticker_df.to_csv(os.path.join(output_dir, f'{symbol}.csv'), index=False)
    print(f"Saved {symbol}.csv with {len(ticker_df)} rows")

print(f"Created {len(df['symbol'].unique())} CSV files in '{output_dir}' folder.") 
```
